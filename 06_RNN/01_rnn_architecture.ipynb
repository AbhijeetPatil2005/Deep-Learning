```python
from keras.layers import Dense,SimpleRNN
from keras import Sequential
```

```python
model =Sequential()

model.add(SimpleRNN(3,input_shape=(4,5)))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```

```python
print(model.get_weights()[0].shape)
model.get_weights()[0]
```
